```python
input ="Hi i am diptanshu. How are u"
output1 = word_tokenize(input)
print(output1)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
input = "this is a second session"
doc = nlp(input)
```

```python
for token in doc:
  print(token)
```

```python
for sent in doc.sents:
  print(sent)
```
